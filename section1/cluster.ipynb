```python
path = '/content/drive/MyDrive/Deep Learning/統計的機械学習　演習/data/'
df = pd.read_csv(path + '3.01. Country clusters.csv')
df
```

```python
plt.scatter(df['Longitude'], df['Latitude'])
plt.xlim(-180, 180)
plt.ylim(-90, 90)
plt.show()
```

```python
x = df.iloc[:,1:3]
kmeans = KMeans(3)
kmeans.fit(x)
```

```python
identified_clusters = kmeans.fit_predict(x)
identified_clusters
```

```python
data_with_clusters = df.copy()
data_with_clusters['Cluster'] = identified_clusters
plt.scatter(data_with_clusters['Longitude'], data_with_clusters['Latitude'], c=data_with_clusters['Cluster'], cmap='rainbow')
plt.show()
```

# sample 2

```python
data = pd.read_csv(path +'3.12. Example.csv')
data.head()
```

```python
plt.scatter(data['Satisfaction'], data['Loyalty'])
plt.show()
```

```python
kmeans = KMeans(2)
data_sat = data.copy()
x = data_sat.iloc[:,0:2]
kmeans.fit(x)
kmeans_predict = kmeans.fit_predict(x)
```

```python
data_sat['Cluster'] = kmeans_predict
plt.scatter(data_sat['Satisfaction'], data_sat['Loyalty'], c=data_sat['Cluster'], cmap='rainbow')
plt.show()
```

```python
from sklearn import preprocessing
x_scaled = preprocessing.scale(x)
x_scaled
```

```python
wcss =[]
for i in range(1, 10):
  kmeans = KMeans(i)
  kmeans.fit(x_scaled)
  wcss_iter = kmeans.inertia_
  wcss.append(wcss_iter)
plt.plot(range(1,10), wcss)
plt.show
```

```python
kmeans = KMeans(4)
kmeans.fit(x_scaled)
kmeans_predict = kmeans.fit_predict(x_scaled)
clusters_new = x.copy()
clusters_new['Cluster'] = kmeans_predict
```

```python
plt.scatter(clusters_new['Satisfaction'], clusters_new['Loyalty'], c=clusters_new['Cluster'], cmap='rainbow')
plt.show()
```
